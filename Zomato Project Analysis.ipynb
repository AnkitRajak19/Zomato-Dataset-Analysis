Step:-1 Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step:-2 Create The Data Frame

```python
df=pd.read_csv('/content/drive/MyDrive/Zomato Data/Zomato data.csv')
df
```

#Convert the data type of column-Rate

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

df['rate']=df['rate'].apply(handleRate)
print(df.head())
```

```python
df
```

```python
df.info()
```

#Type Of Resturent

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type Of Restaurant')
```

#Conclusion- majority of the resturent falls in dinning category

```python
df.head()
```

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c='green',marker='o')
plt.xlabel('Type Of Restaurant',c='r',size=15)
plt.ylabel('Total Votes',c='r',size=15)
```

#Conclusion- Dinning resturent has recieved maximum votes

```python
df.head()
```

```python
plt.hist(df['rate'],bins=5)
plt.title('Rating Distribution')
plt.show
```

#Conclusion-The majority resturants received ratings from 3.5 to 4.

#Average order spending by couples

```python
df.head()
```

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

#conclusion-the majority of couples prefer resturants with an approximatly cost of 300 RS.

#Which mode receives maximum rating

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df)
```

#Conclusion- Offline order received lower rating in comparison to online order

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['online_order'],hue=df['rate'])
```

```python
df.head()
```

```python
pivot_table=df.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap='YlGnBu',fmt='d')
plt.title('heatmap')
plt.xlabel('Online Order')
plt.ylabel('Type Of Restaurant')
plt.show()
```
